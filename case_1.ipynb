```python
x_train,x_val,y_train,y_val=train_test_split(x_train,y_train,test_size=368/1102,random_state=0)
```

```python
x_train.shape
```

```python
# tress estimators

num_trees=[10,20,25,50,75,100,200,500,750,1000]
```

```python
for tree in num_trees:
    rf=RandomForestClassifier(n_estimators=tree)
    rf.fit(x_train,y_train)
    print('trees =',tree,': auc is ',metrics.roc_auc_score(y_val,rf.predict_proba(x_val)[:,1]))
```

```python
# now finally on test  # above roc is greatest on tree=100
rf1=RandomForestClassifier(n_estimators=100)
rf1.fit(x_train,y_train)
print('roc_auc is ',metrics.roc_auc_score(y_test,rf1.predict_proba(x_test)[:,1]))
```
